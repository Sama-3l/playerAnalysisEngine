```python
import pandas as pd
from IPython.display import display
df = pd.read_csv("vlrscrape1.3.csv", parse_dates=["Date"])
df
```

```python
df['Agent'] = df['Agent'].str.replace('-', 'Deadlock')
df
```

```python
desired_rows = df[df.apply(lambda row: row.astype(str).str.contains('Paper Rex').any(), axis=1)]
desired_rows
```

```python
teams = {}
for team in df['Team'].unique():
    teams[team] = {}

empty_df = pd.DataFrame()

for index, row in df.iterrows():
    column_value = row['Team']
    # Do something with the column value
    if teams[row['Team']].get(row['Player']) is None:
        teams[row['Team']][row['Player']] = {}

teams
```

```python
for key in teams.keys():
    for player in teams[key].keys():
        teams[key][player]['winStats'] = df[(df['Winner'] == key) & (df['Player'] == player) & ~df.apply(lambda row: row.astype(str).str.contains('Deadlock').any(), axis=1)]

for key in teams.keys():
    for player in teams[key].keys():
        print(player)
        display(teams[key][player]['winStats'])
```

```python
selected_columns = df[['Player', 'D', 'FK', 'Team']]
all_teams = df['Team'].unique()
for eachTeam in all_teams:
    selected_columns = selected_columns[~selected_columns.apply(lambda row: row.astype(str).str.contains('-').any(), axis=1) & (selected_columns['Team'] == eachTeam)]
    display(selected_columns)
```

```python
#Seperates  the databases of all players according to teams as the following JSON structure
'''
teams = {
    "team-name" : {
        'player-name' : {
            'compiled-stats': <DataFrame with averaged out stats>
            'allStats': <DataFrame more details>
        }
    }
}
'''

for key in teams.keys():
    for player in teams[key].keys():
        teams[key][player]['compiledStats'] = df[df.apply(lambda row: row.astype(str).str.contains(player).any(), axis=1) & df.apply(lambda row: row.astype(str).str.contains('Deadlock').any(), axis=1)]
        teams[key][player]['allStats'] = df[df.apply(lambda row: row.astype(str).str.contains(player).any(), axis=1) & ~df.apply(lambda row: row.astype(str).str.contains('Deadlock').any(), axis=1)]

for key in teams.keys():
    for player in teams[key].keys():
        print(player)
        display(teams[key][player]['compiledStats'])
```
